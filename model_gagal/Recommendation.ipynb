```python
import pandas as pd

df_user = pd.read_csv("dataset/MOCK_DATA_USER.csv")
df_user.head()
```

```python
df_user["personalization"] = df_user["personalization1"] + ", " + df_user["personalizatoin2"] + ", " + df_user["personalization3"] + ", " + df_user["personalization4"] + ", " + df_user["personalization5"] + ", " + df_user["personalization6"]
df_user = df_user.drop(["personalization1", "personalizatoin2", "personalization3", "personalization4", "personalization5", "personalization6"], axis=1)
df_user.head()
```

```python
df_cafe = pd.read_csv("dataset/New_CafeJakarta.csv")
df_cafe = df_cafe.rename(columns={"name":"cafe_name"})
df_cafe = df_cafe.drop(["description", "website", "featured_image", "main_category", "categories", "workday_timing", "closed_on",
                        "phone", "review_keywords", "link"], axis=1)
df_cafe = df_cafe.drop(['outdoor', 'Indoor',
       'RetroVintage', 'MinimalisCafe', 'IndustrialCafe', 'ModernCafe',
       'ArtCafe', 'PetCafe', 'BooksCafe', '24HoursCafe', 'MeetingCafe',
       'StudyCafe', 'goodViews', 'FamilyCafe', 'CountryFood', 'Smoking',
       'NonSmoking', 'Coffee', 'NonCoffee', 'GardenCafe'], axis=1)
df_cafe.head()
```

```python
df_cafe.to_csv('MOCK_DATA_CAFE.csv', index=False)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Preprocess the dataframes
user_features = pd.get_dummies(df_user['personalization'].str.split(', ', expand=True))
# Get the list of all features
all_features = set(df_user['personalization'].str.split(', ', expand=True).stack().unique()).union(
    set(df_cafe['feature'].str.split(', ', expand=True).stack().unique()))

# Add dummy columns for missing features
for feature in all_features:
    if feature not in df_user.columns:
        df_user[feature] = 0

cafe_features = pd.get_dummies(df_cafe['feature'].str.split(', ', expand=True))
```

```python
# Compute the similarity matrix
similarity_matrix = cosine_similarity(user_features, cafe_features)
```
