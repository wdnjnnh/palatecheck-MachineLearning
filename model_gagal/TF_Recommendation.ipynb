```python
import pandas as pd

df = pd.read_csv('dataset/Dataset_PalatePicks.csv')
df.head(10)
```

```python
kolom_float = ['outdoor', 'Indoor', 'RetroVintage', 'MinimalisCafe', 'IndustrialCafe', 'ModernCafe', 'ArtCafe', 'PetCafe', 'BooksCafe', 
              '24HoursCafe', 'MeetingCafe', 'StudyCafe', 'goodViews', 'FamilyCafe', 'CountryFood', 'Smoking', 'NonSmoking', 'Coffee', 
              'NonCoffee', 'GardenCafe', 'Rating']

df[kolom_float] = df[kolom_float].fillna(0)

# Mengubah data float menjadi int
df[kolom_float] = df[kolom_float].astype(int)

df.head(5)
```

```python
df = df.rename(columns={'Name':'User_Name', 'name':'cafe_name'})
```

```python
df.columns
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import tensorflow as tf
from tensorflow.keras.layers import Input, Embedding, Dot, Flatten
from tensorflow.keras.models import Model
```

```python
# Mengubah kolom-kolom yang berisi nilai 0 dan 1 menjadi tipe data boolean
df[['Kpopers', 'JapanLovers', 'AnimalLovers', 'Quite', 'MusicLovers', 'BookLovers', 'ArtLovers', 'ViewsLovers', 
    'CoffeeLovers', 'NonCoffeeLovers', 'groupsComer']] = df[['Kpopers', 'JapanLovers', 'AnimalLovers', 'Quite', 
                                                             'MusicLovers', 'BookLovers', 'ArtLovers', 'ViewsLovers', 
                                                             'CoffeeLovers', 'NonCoffeeLovers', 'groupsComer']].astype(bool)
```

```python
# Menghitung jumlah pengguna dan tipe cafe
n_users, n_cafes = len(df['User_id'].unique()), len(df['cafe_id'].unique())
```

```python
# Membuat model
user_input = Input(shape=[1])
cafe_input = Input(shape=[1])
user_embedding = Embedding(n_users, 10)(user_input)
cafe_embedding = Embedding(n_cafes, 10)(cafe_input)
user_vec = Flatten()(user_embedding)
cafe_vec = Flatten()(cafe_embedding)
prod = Dot(axes=1)([user_vec, cafe_vec])
model = Model([user_input, cafe_input], prod)
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Membuat array rating
X = df[['User_Name', 'outdoor', 'Indoor', 'RetroVintage', 'MinimalisCafe', 'IndustrialCafe', 'ModernCafe', 
        'ArtCafe', 'PetCafe', 'BooksCafe', '24HoursCafe', 'MeetingCafe', 'StudyCafe', 'goodViews', 'FamilyCafe', 
        'CountryFood', 'Smoking', 'NonSmoking', 'Coffee', 'NonCoffee', 'GardenCafe']].values
y = df['Rating'].values
```

```python
# Melatih model
model.fit(x=[X[:,0], X[:,1:21]], y=y, epochs=10)
```

```python
# Fungsi untuk merekomendasikan tipe cafe berdasarkan preferensi pengguna
def recommend_cafe(user_name, df, model):
    # Mengambil indeks pengguna
    user_index = df[df['Name'] == user_name].index[0]

    # Mengambil preferensi pengguna
    user_preferences = df.iloc[user_index][['outdoor', 'Indoor', 'RetroVintage', 'MinimalisCafe', 'IndustrialCafe', 'ModernCafe', 'ArtCafe', 'PetCafe', 'BooksCafe', '24HoursCafe', 'MeetingCafe', 'StudyCafe', 'goodViews', 'FamilyCafe', 'CountryFood', 'Smoking', 'NonSmoking', 'Coffee', 'NonCoffee', 'GardenCafe']].values.reshape(1,-1)

    # Menghitung rating untuk setiap tipe cafe
    ratings = model.predict([np.array([user_index]*n_cafes), np.arange(n_cafes)])

    # Mengambil tipe cafe yang disukai oleh pengguna saat ini
    user_preferences = df.iloc[user_index][['outdoor', 'Indoor', 'RetroVintage', 'MinimalisCafe', 'IndustrialCafe', 'ModernCafe', 'ArtCafe', 'PetCafe', 'BooksCafe', '24HoursCafe', 'MeetingCafe', 'StudyCafe', 'goodViews', 'FamilyCafe', 'CountryFood', 'Smoking', 'NonSmoking', 'Coffee', 'NonCoffee', 'GardenCafe']]

    # Mengambil tipe cafe yang direkomendasikan
    recommended_cafe = df.columns[1:][ratings[user_index]>ratings[user_index].mean()]

    # Mengembalikan tipe cafe yang direkomendasikan
    return recommended_cafe
```

```python
# Contoh penggunaan fungsi recommend_cafe
user_name = 'John'
recommended_cafe = recommend_cafe(user_name, df, model)
print("Tipe cafe yang direkomendasikan untuk", user_name, "adalah:", recommended_cafe)
```
