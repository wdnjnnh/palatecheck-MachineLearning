```python
from sklearn.metrics.pairwise import cosine_similarity

# Menghitung kemiripan antara pengguna
user_similarity = cosine_similarity(df_user.drop(['Email', 'Password', 'Name', 'Gender', 'age '], axis=1))
```

```python
import numpy as np

# Fungsi untuk merekomendasikan tipe cafe berdasarkan preferensi pengguna
def recommend_cafe(user_id, df_user, user_similarity):
    # Mengambil indeks pengguna
    user_index = df_user[df_user['User_id'] == user_id].index[0]

    # Menghitung kemiripan antara pengguna lain dan pengguna saat ini
    similarity_scores = list(enumerate(user_similarity[user_index]))

    # Mengurutkan pengguna berdasarkan kemiripan
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    # Mengambil indeks pengguna yang paling mirip
    most_similar_user_index = similarity_scores[1][0]

    # Mengambil preferensi pengguna yang paling mirip
    most_similar_user_preferences = df_user.iloc[most_similar_user_index][['Kpopers', 'JapanLovers', 'AnimalLovers', 'Quite', 
        'MusicLovers', 'BookLovers', 'ArtLovers', 'ViewsLovers', 
        'CoffeeLovers', 'NonCoffeeLovers', 'groupsComer']]

    # Mengambil preferensi pengguna saat ini
    user_preferences = df_user.iloc[user_index][['Kpopers', 'JapanLovers', 'AnimalLovers', 'Quite', 
        'MusicLovers', 'BookLovers', 'ArtLovers', 'ViewsLovers', 
        'CoffeeLovers', 'NonCoffeeLovers', 'groupsComer']]

    # Menghitung perbedaan antara preferensi pengguna saat ini dan pengguna yang paling mirip
    difference = np.subtract(most_similar_user_preferences, user_preferences)

    # Mengambil tipe cafe yang disukai oleh pengguna yang paling mirip dan tidak disukai oleh pengguna saat ini
    recommended_cafe = difference[difference > 0].index.tolist()

    # Mengembalikan tipe cafe yang direkomendasikan
    return recommended_cafe
```

```python
# Contoh penggunaan fungsi recommend_cafe
user_id = 1
recommended_cafe = recommend_cafe(user_id, df_user, user_similarity)
print("Tipe cafe yang direkomendasikan untuk user_id", user_id, "adalah:", recommended_cafe)
```
